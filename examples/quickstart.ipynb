## 1. Name-Based Classification

The first stage classifies gender from the first name using `gender_guesser`.
Ambiguous names (andy, unknown, mostly_male, mostly_female, cross-cultural Italian) are flagged.

```python
from genderphoto import classify_name

# Unambiguous: James in US -> male
print(classify_name('James', 'US'))

# Ambiguous: Andrea in US -> cross-cultural Italian name
print(classify_name('Andrea', 'US'))

# Not ambiguous: Andrea in Italy -> male
print(classify_name('Andrea', 'IT'))
```

## 2. Single Inventor (Full Pipeline)

For ambiguous names, the pipeline searches photos and runs the DeepFace + VLM ensemble.

**Note:** This requires internet access (for Bing image search) and optionally Ollama (for VLM fallback).

```python
from genderphoto import classify_inventor

result = classify_inventor(
    name='Andrea Cavalleri',
    affiliation='Max Planck Hamburg',
    country_code='DE',
    save_photo_flag=True,
)

print(f"Gender: {result['gender']}")
print(f"Method: {result['method']}")
print(f"Confidence: {result['confidence']}")
print(f"Photo saved: {result['photo_saved_path']}")
```

## 3. Batch Processing

Process a DataFrame of inventors. Only ambiguous names trigger the photo pipeline.

```python
import pandas as pd
from genderphoto import classify_batch

df = pd.DataFrame([
    {'inventor_name': 'James Heckman', 'affiliation': 'University of Chicago', 'country_code': 'US'},
    {'inventor_name': 'Andrea Cavalleri', 'affiliation': 'Max Planck Hamburg', 'country_code': 'DE'},
    {'inventor_name': 'Jennifer Doudna', 'affiliation': 'UC Berkeley', 'country_code': 'US'},
    {'inventor_name': 'Fei-Fei Li', 'affiliation': 'Stanford University', 'country_code': 'US'},
])

result_df = classify_batch(
    df,
    save_photos=True,
    photo_dir='./inventor_photos',
    checkpoint_path='./results/checkpoint.csv',
)

result_df[['inventor_name', 'gender_final', 'gender_method', 'is_ambiguous']]
```
